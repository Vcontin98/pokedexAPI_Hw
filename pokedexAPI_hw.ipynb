```python
import requests
import pprint
from IPython.display import clear_output as co
pp = pprint.PrettyPrinter(indent=4)


class Pokemon:
    def __init__(self, name=None, abilities=None, types=None, height=None, weight=None):
        self.name = name
        if abilities is None:
            self.abilities = []
        else:
            self.abilities = abilities
        if types is None:
            self.types = []
        else:
            self.types = types
        self.height = height
        self.weight = weight

    def from_dict(self, data):
        for field in ['name', 'abilities', 'types', 'height', 'weight']:
            if field in data:
                
                setattr(self, field, data[field])

    def __repr__(self):
        return f'<Pokemon: {self.name}>'

    def __str__(self):
        return f'{self.name}'
class Pokedex:
    _list = []

    @classmethod
    def starting(self):
        print("""Type 'show' to view Pokedex.
Type 'quit' to exit the Pokedex.
Type 'sort' to view a categorized version of your Pokedex.""")
    @classmethod
    def add(self, name):
        
        if Pokedex._list:
            for p in Pokedex._list:
                if name.title() == p.name:
                    input('Pokemon already exists. Try another one. ')
                    return
        try:
            
            print('Loading Please wait.')
            res = requests.get(
                f'https://pokeapi.co/api/v2/pokemon/{name.lower()}').json()
            p = Pokemon()
            
            data = {
                'name': res['name'].title(),
                'abilities': [a['ability']['name'].title() for a in res['abilities']],
                'types': [t['type']['name'].title() for t in res['types']],
                'height': res['height'],
                'weight': res['weight']
            }
            p.from_dict(data)
            
            Pokedex._list.append(p)
        except:
            input('Error. Try again. ')

    @classmethod
    def show(self):
        return Pokedex.all()

    @classmethod
    def run(self):
        end = False
        input('Welcome to Pokemon Pokedex! Press any key to continue. ')
        while not end:
            co()
            self.starting()
            name = input(
                'Type the name of a Pokemon to add. ').lower()
            if name == 'quit':
                end = True
                break
            elif name == 'show':
                Pokedex.show()
                input('Press any key to continue. ')
            elif name == 'sort':
                Pokedex.poke()
                input('Press any key to continue. ')
            else:
                Pokedex.add(name)

    @classmethod
    def poke(self):
        
        poke_dict = {}
        for p in Pokedex._list:
            for t in p.types:
                if t not in poke_dict:
                    poke_dict[t] = {}
        for p in Pokedex._list:
            for t in p.types:
                
                if p.name not in poke_dict[t]:
                    poke_data = {
                        p.name: {
                            'abilities': p.abilities,
                            'height': p.height,
                            'weight': p.weight
                        }
                    }
                    poke_dict[t].update(poke_data)
                else:
                    print('Pokemon already exists')
        pp.pprint(poke_dict)

    @classmethod
    def all(self):
        if Pokedex._list:
            for i in Pokedex._list:
                print(i)
Pokedex.run()
```
